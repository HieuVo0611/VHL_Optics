```python
save_path='D:/Work/Water Environment Pollution/folder/doubledataset/Image/'
```

```python
# Tạo thư mục đích nếu chưa tồn tại
os.makedirs(save_path, exist_ok=True)

# Kiểm tra và sao chép hình ảnh
for image_name in metadata2['ID_img']:
    image_name_lst = str(image_name).split('.thumbnail')
    if len(image_name_lst) ==2:
        image_name = image_name_lst[0]+'.jpg'
    save_image_path = os.path.join(save_path, image_name)
    
    # Kiểm tra nếu ảnh đã có trong thư mục đích
    if os.path.exists(save_image_path):
        print(f"Image already exists in save folder: {image_name}")
        continue
    
    found = False
    for path in lst_path:
        image_path = os.path.join(path, image_name)
        if os.path.exists(image_path):
            shutil.copy(image_path, save_path)
            print(f"Copied: {image_name} to {save_path}")
            found = True
            break
    if not found:
        print(f"Image not found: {image_name}")
```
